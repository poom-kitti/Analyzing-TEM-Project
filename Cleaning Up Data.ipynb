There seems to be extra columns with no values, so I will first drop all those columns

```python
df = df.dropna(axis=1, how='all')
```

## Processing Data into Readable Table

```python
#Changing time to proper timestamp
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
```

#### Altering score
- Since scoring is from 1 to 5, I will normalize the scores to 0 to 5
- Question 2 is a negative question, so a score of 1 originally means the best score, while the score of 5 is the worst score

```python
#Getting a list of column names based on the question type
q1 = ['P'+str(x)+'Q1' for x in range(1,5)]
q2 = ['P'+str(x)+'Q2' for x in range(1,5)]
q3 = ['P'+str(x)+'Q3' for x in range(1,5)]
```

```python
df[q1+q3]=(df[q1+q3]-1)*5/4
```

```python
df[q2] = (5-df[q2])*5/4
```

```python
df.head()
```

#### Creating a new clean table
- I will also add an average socre for each part

```python
df_clean = df.copy()
```

```python
#Getting list of column names based on part
p1 = ['P1Q'+str(x) for x in range(1,4)]
p2 = ['P2Q'+str(x) for x in range(1,4)]
p3 = ['P3Q'+str(x) for x in range(1,4)]
p4 = ['P4Q'+str(x) for x in range(1,4)]
```

```python
#Create a column for average score based on part
for index, part in enumerate([p1,p2,p3,p4]):
    df_clean['P'+str(index+1)+' Overall'] = df_clean[part].sum(axis=1)/3
```

```python
df_clean.head()
```

```python
#Formatting the DataFrame to how I want to present them
cols = df_clean.columns.tolist()
new_cols = cols[:5] + cols[-4:-3] + cols[5:8] + cols[-3:-2] + cols[8:11] + cols[-2:-1] + cols[11:14] + cols[-1:] +cols[14:16]
```

```python
df_clean = df_clean[new_cols]
```

```python
df_clean.head()
```

```python
#Turn the clean DataFrame to csv file
df_clean.to_csv('tem_survey_clean.csv', encoding='utf-8', index=False)
```

## Processing Data into Table for Tableau

```python
#Creating a new DataFrame with information I wanted
df_tableau = pd.DataFrame(columns=['Part', 'Question', 'Score', 'Type'])
```

```python
#A method on type of user based on the overall score
def user_type(x):
    if x > 8: return 'Promoter'
    elif x > 6: return 'Passive'
    else: return 'Detractor'
```

```python
#I want to somehow separate score rated by each user based on part and question
df_clean.T.head()
```

```python
#Getting what type the user is
type_df = df['Overall'].apply(user_type)
type_df
```

```python
for i in range(1,5):
    for j in range(1,4):
        #Getting DataFrame of a score rated by user for specific question of a certain part & the user type
        temp_df = pd.concat([type_df,df_clean.T.loc['P'+str(i)+'Q'+str(j)]], axis=1)
        #Filling in values of the temporary DataFrame to match the question and part of the score
        temp_df['Part'] = 'Part ' + str(i) 
        temp_df['Question'] = 'Question ' + str(j)
        #Formating the temporary DataFrame
        temp_df.columns = ['Type', 'Score', 'Part', 'Question']
        temp_df = temp_df[['Part', 'Question', 'Score', 'Type']]
        #Concatenate the new DataFrame to the DataFrame I want
        df_tableau = pd.concat([df_tableau, temp_df], axis=0, ignore_index=True)
```

```python
df_tableau
```

```python
#Turn the DataFrame to a csv file
df_tableau.to_csv('tem_survey_clean_by_part_and_question.csv', encoding='utf-8', index=False)
```
